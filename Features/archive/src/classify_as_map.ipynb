```python
def updateMap(Maps):
    Index = [433,434,459,460,461,462,463,464,487,488,489,490,491,492,515,516,517,518,519,520]
    Map_temp = list(Maps)
    for i in Index:
        Map_temp[i] = " "
    Map = "".join(Map_temp)
    return Map
```
